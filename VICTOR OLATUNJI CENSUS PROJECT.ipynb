## EMPLOYMENT CATEGORY
- This particular new column is added to simplify employment
- Function to Make a new column named as "Employement Category"

```python
def employment_status(x):
    if (x["Occupation"].lower() == "child"):
        return "Child"
    elif (x["Occupation"].lower() == "student") & (x["Age"] <18):
        return "Student (Child)"
    elif (x["Occupation"].lower() == "student") or ("student" in x["Occupation"].lower()):
        return "Student"
    elif (x["Occupation"].lower() == "unemployed"):
        return "Unemployed"
    elif ("retired" in x["Occupation"].lower()) or ("retire" in x["Occupation"].lower()):
        return "Retired"
    else:
        return "Employed"
census_df["Employment Status"]=census_df.apply(lambda x: employment_status(x),axis=1)
```

```python
census_df["Employment Status"].value_counts()
```

## HOUSEHOLD OCCUPANCY COUNTS
- This indicates the number of individual per household
- This value can not be sum together to get the total occupancy

```python
census_df["Household Occupancy"] = census_df.groupby(["House Number","Street"])["Street"].transform("count")
# census_df["Household Occupancy"] = census_df.groupby(["House Number","Street"]).transform("count")
census_df.head()
```

```python
male_census_ages = census_df[census_df['Gender'] == 'Male']['Age'].to_list()

female_census_ages = census_df[census_df['Gender'] == 'Female']['Age'].to_list()

age_intervals = []
male_count = []
female_count = []
next_start = 0
for i in range(0,112,5):
   interval = f'{next_start}-{next_start + 4}'
   next_start=next_start+4+1
   age_intervals.append(interval)
   male_count.append(0)
   female_count.append(0)


def get_age_group(age):
   index = int(age/5)
   return age_intervals[index]

census_df['age_group'] = census_df['Age'].apply(get_age_group)




for age in male_census_ages:
   index = int(age/5)
   male_count[index] = male_count[index] + 1


for age in female_census_ages:
   index = int(age/5)
   female_count[index] = female_count[index] + 1

age_pyramid_df = pd.DataFrame({
   'Age' : age_intervals,
   'Female' : female_count,
   'Male' : [-1 * val for val in male_count]
})

plt.figure(figsize=(12,8))
age_pyramid = sns.barplot(x='Male', y='Age', data=age_pyramid_df, order=reversed(age_intervals), color=('yellow'), label='Male')
age_pyramid = sns.barplot(x='Female', y='Age', data=age_pyramid_df, order=reversed(age_intervals), color=('red'), label='Female')
age_pyramid.legend()
plt.title('Age Pyramid')
age_pyramid.set(xlabel='Population Count', ylabel='Age Group')
```
